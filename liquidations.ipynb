```python
import csv
import pandas as pd

def load_data(filename):
    reader = csv.DictReader(open(filename, 'r'))
    rows = []
    for row in reader:
        float_keys = [
            'price',
            'market_cap',
            'total_volume',
            '1d_ta',
            '7d_ta',
            '30d_ta',
        ]
        for key in float_keys:
            if key not in row:
                continue
            if not row[key]:
                row[key] = 0
            else:
                row[key] = float(row[key])
        rows.append(row)
    return rows

v2_rows = load_data('aave_v2_full.csv')
v3_rows = load_data('liq_aave_v3.csv')
df = pd.json_normalize(v2_rows)
weth = '0xc02aaa39b223fe8d0a0e5c4f27ead9083c756cc2'
df_weth = df[df['collateralAsset'].eq(weth)]
df_weth
```

```python
# get data in following format:
# [liquidation_date, price_at_date]

eth_usd_rows = load_data('eth_usd.csv')
eth_usd_df = pd.json_normalize(eth_usd_rows)

def price_on_date(date):
    for row in eth_usd_rows:
        if row['date'] == date:
            return row['price']
    return None

def days_to_recover(date, price):
    count = 0
    for row in eth_usd_rows:
        if row['date'] <= date:
            continue
        count += 1
        if row['price'] > price:
            return count
    return 0

liquidation_rows = []
for row in v2_rows:
    if row['collateralAsset'] != weth:
        continue
    date = row['evt_block_time'].split(' ')[0]
    price = price_on_date(date)
    days = days_to_recover(date, price)
    row = {
        'date': date,
        'price': price,
        'days_to_recover': days,
    }
    liquidation_rows.append(row)
    
df = pd.json_normalize(liquidation_rows)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

never_recovered = df[df['days_to_recover'].eq(0) | df['days_to_recover'].gt(100) ]
above_today = df[df['price'].lt(2500)]
recovered = df[df['days_to_recover'].gt(0)]
recovered = recovered[recovered['days_to_recover'].lt(100)]
bins = np.arange(0, recovered['days_to_recover'].max(), 10)
fig, ax = plt.subplots()
plt.hist(recovered['days_to_recover'], bins=bins)
ax.set_title('Days for liquidated position to recover')
plt.show()
print('Num recovered: %s' % len(recovered))
print('Num did not recovered: %s' % len(never_recovered))
print('Percent recovered: %0.1f%%' % (100 * len(recovered) / len(df)))
print('Above today: %s' % len(above_today))
```
