```python
import csv
import pandas as pd

def load_data(filename):
    reader = csv.DictReader(open(filename, 'r'))
    rows = []
    for row in reader:
        float_keys = [
            'price',
            'market_cap',
            'total_volume',
            '1d_ta',
            '7d_ta',
            '30d_ta',
        ]
        for key in float_keys:
            if key not in row:
                continue
            if not row[key]:
                row[key] = 0
            else:
                row[key] = float(row[key])
        rows.append(row)
    return rows

v2_rows = load_data('aave_v2_full.csv')
v3_rows = load_data('liq_aave_v3.csv')
df = pd.json_normalize(v2_rows)
weth = '0xc02aaa39b223fe8d0a0e5c4f27ead9083c756cc2'
df_weth = df[df['collateralAsset'].eq(weth)]
df_weth
```

```python
# get data in following format:
# [liquidation_date, price_at_date]

eth_usd_rows = load_data('eth_usd.csv')
eth_usd_df = pd.json_normalize(eth_usd_rows)

def price_on_date(date):
    for row in eth_usd_rows:
        if row['date'] == date:
            return row['price']
    return None

def days_to_recover(date, price):
    count = 0
    for row in eth_usd_rows:
        if row['date'] <= date:
            continue
        count += 1
#         if count < 5:
#             continue
        if row['price'] > price:
            return count
    return 0

liquidation_rows = []
for row in v2_rows:
    if row['collateralAsset'] != weth:
        continue
    date = row['evt_block_time'].split(' ')[0]
    price = price_on_date(date)
    days = days_to_recover(date, price)
    row = {
        'date': date,
        'price': price,
        'days_to_recover': days,
    }
    liquidation_rows.append(row)
    
df = pd.json_normalize(liquidation_rows)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

never_recovered = df[df['days_to_recover'].eq(0) | df['days_to_recover'].gt(100) ]
above_today = df[df['price'].lt(2500)]
recovered = df[df['days_to_recover'].gt(0)]
recovered = recovered[recovered['days_to_recover'].lt(100)]
bins = np.arange(0, recovered['days_to_recover'].max(), 5)
fig, ax = plt.subplots()
plt.hist(recovered['days_to_recover'], bins=bins)
ax.set_title('Days for liquidated position to recover')
plt.show()
print('Num recovered: %s' % len(recovered))
print('Num did not recovered: %s' % len(never_recovered))
print('Percent recovered: %0.1f%%' % (100 * len(recovered) / len(df)))
print('Above today: %s' % len(above_today))
```

```python
lq_rows = load_data('liquity_liquidations.csv')
lq_df = pd.json_normalize(lq_rows)
lq_df['value'] = lq_df['collateral'].astype(float) * lq_df['price'].astype(float)
lq_df
```

```python
import matplotlib

def did_recover(date, price, price_rows, recover_buffer):
    for row in price_rows:
        if row['date'] > date and row['price'] > price * (1 + recover_buffer):
            return (True, row)
    #print('did not recover:', date, price )
    return False, None

def num_recovered(df, price_rows, recover_buffer):
    count = 0
    for row in range(len(df)):
        price, timestamp = df.at[row, 'price'], df.at[row, 'timestamp']
        date = str(timestamp).split(' ')[0]
        ok, recover_row = did_recover(date, price, price_rows, recover_buffer)
        if ok:
            count += 1
    return (count, count / len(lq_df))


points = []
for i in range(100):
    buffer = i / 100
    (count, percent) = num_recovered(lq_df, eth_usd_rows, buffer)
    if i % 10 == 0:
        print(buffer, percent)
    points.append((buffer, percent))

(count, percent) = num_recovered(lq_df, eth_usd_rows, 0.01)
print('%s recoverd out of %s (%.1f%%)' % (count, len(lq_df), percent * 100))

x = [p[0] for p in points]
y = [p[1] for p in points]

fig, ax = plt.subplots()
plt.scatter(x, y, color='blue', marker='o')

ax.set_title('Liquidation recovery with various buffers')
ax.set_ylabel('% of positions that recover')
ax.set_xlabel('% price recover above liquidation price')

ax.get_xaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(
        lambda x, p: '%.0f%%' % (x*100)
    )
)
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(
        lambda x, p: '%.0f%%' % (x*100)
    )
)

plt.ylim(0, 1.05)

# Displaying the plot
plt.show()
```
